# Handling missing Values

```python
filter_data.isna().mean().round(4) *100
# this code gives the peercebtage of missing values in each column
```

# replacing the missing values with mean in numerical columns

```python
filter_data=filter_data.fillna(filter_data.mean())
```

```python
filter_data.isna().sum()
```

# tests_units column is categorical. since from mode we know the mode of this column this column is "samples_tested", so lets fill the null values in this column with mode value.

```python
filter_data =filter_data.fillna("samples tested")
```

```python
filter_data.isna().sum()
```

# Convert date column to ordinal

```python
import datetime as dt
filter_data["date"]=pd.to_datetime(filter_data["date"])
filter_data["date"]=filter_data["date"].map(dt.datetime.toordinal)
```

# drop all categorical columns

```python
## dropping all categorical values
filter_data = filter_data.drop(["iso_code","continent","location","tests_units","population","population_density","median_age",
                                              "aged_65_older","aged_70_older","gdp_per_capita","extreme_poverty","cvd_death_rate","diabetes_prevalence",
                                              "female_smokers","male_smokers","handwashing_facilities","hospital_beds_per_thousand","life_expectancy"], axis = 1)
```

```python
filter_data
```

# Selecting total cases as target variable and other columns as features
### X cotains the feature variable and Y contains the target variable

```python
# selecting total cases as target variable and other features as undependent variable
X = filter_data.drop("total_cases", axis = 1)
Y = filter_data["total_cases"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.33, random_state = 50)
```

```python
X_train.shape, X_test.shape
```

# Linear Regression


```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

```python
Y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(Y_test, Y_pred)
score
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
```

```python
rf.fit(X_train, Y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
y_pred, Y_test
```

```python
score = r2_score(Y_test, y_pred)
score
```
